```python
reference_image_path = "/content/drive/MyDrive/Newdata/Adino/WP_20200823_112.jpg"
reference = cv2.imread(reference_image_path)
```

```python
path = "/content/drive/MyDrive/Newdata"
labels = ["Adino", "SCC", "SIL"]
```

```python
dest_path = "/content/drive/MyDrive/colorNormalized"
```

```python
image_data,label_data = [],[]
print("start reading")
for l in labels:
    data_path = os.path.join(path,l)
    dest_folder = os.path.join(dest_path,l)
    if not(os.path.isdir(dest_folder)):
      os.mkdir(dest_folder)
    for f in os.listdir(data_path):
        image = cv2.imread(os.path.join(data_path, f))
        matched = match_histograms(image, reference, multichannel= True)
        dest_file_path = os.path.join(dest_folder,f)
        cv2.imwrite(dest_file_path, matched)
        # matched = cv2.resize(matched, (144,144))
        # image_data.append(matched)
        # label_data.append(labels.index(l)) ## only taking index of the label (Eg. 1 for Normal, 2 for Precancer ...)
print("Finished")
```
